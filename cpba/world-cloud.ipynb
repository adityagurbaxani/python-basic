```python
# Imports
import wikipedia
import re
import matplotlib.pyplot as plt
```

```python
# Specify the title of the Wikipedia page
wiki = wikipedia.page('Cryptocurrency')
# Extract the plain text content of the page
text = wiki.content
# Clean text
text = re.sub(r'==.*?==+','',text)
text = text.replace('\n', '')
```

```python
def plot_cloud(wordCloud):
    # Set figure size
    plt.figure(figsize=(40, 30))
    # Display image
    plt.imshow(wordCloud)
    # No axis details
    plt.axis("off")
    plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
# Generate word cloud
wc = WordCloud(width = 3000, height = 2000, random_state=1, background_color='aliceblue',
               colormap='tab10', collocations=False, stopwords = STOPWORDS).generate(text)
# Plot
plot_cloud(wc)
```
